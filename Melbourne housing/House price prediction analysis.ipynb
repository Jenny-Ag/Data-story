```python
df1 = df.rename(columns= {'Lattitude' : 'Latitude', 'Longtitude' : 'Longitude'})
df1.info()
```

```python
'''Deleting redundant columns like address, postcode, latitute, longitude, regionname, propertycount, all tell about
location of the house which is preserved in suburb as well, Date of sell/purchase, seller name will not directly 
affect the price so those can be removed, they can be grouped as well into year or period '''
df2 = df1.drop(['Address','Date', 'Method', 'SellerG', 'Postcode', 'Regionname', 'Propertycount','Latitude', 'Longitude'], axis=1)
df2.info()
```

```python
#checking null values for Bedroom2 column
bedzero = df2[df2.Bedroom2.isnull()]
bedzero
```

```python
# null values of bedroom2 column are also null for bathroom, car, landsize, buildingarea, so dropping all values with bedroomas null
df2= df2[df2.Bedroom2.notnull()]
df2.info()
```

```python
# checking the null values for landsize
landsizezero = df2[df2.Landsize.isnull()]
landsizezero
```

```python
# removing only the data where landsize and building area and year built all are zero
df2= df2.drop(df2[(df2.Landsize.isnull()) & (df2.BuildingArea.isnull()) & (df2.YearBuilt.isnull())].index)
df2.info()
```

```python
#removing the observations where price is null
df2= df2.drop(df2[(df2.Price.isnull())].index)
df2.info()
```

```python
# checking the observations where Buildingarea is null
Bldareazero = df2[df2.BuildingArea.isnull()]
Bldareazero
```

```python
'''There are lot of observations almost 6000 out of 16000 where Building ares is not given, so instead of 
dropping all the observations where building area is zero, for the time being I would drop that columns for model building purposes.'''
df2 = df2.drop(['BuildingArea'], axis=1)
df2.info()
```

```python
'''Similarly for year built there are aroundt 6000 out of 16000 where year is not given, so instead of 
dropping all the observations where year is not given, for the time being I would drop that columns for model building purposes.'''
df2 = df2.drop(['YearBuilt'], axis=1)
df2.info()

```

```python
# checking for observations with Bathroom nan
bathzero = df2[df2.Bathroom.isnull()]
bathzero
```

```python
# checking for the observations with car nan
carzero = df2[df2.Car.isnull()]
carzero
```

```python
# Both Car nan and bath nan and Landsize nan are not too many observations so dropping them for now is good idea
df2 = df2.drop(df2[(df2.Bathroom.isnull()) | (df2.Car.isnull()) | (df2.Landsize.isnull())].index)
df2.info()
```

### converting non numerical variables into variables using one-hot encoding : One-hot encoding categorical variables is a best-practice though, it allows you to check for underrepresented categories and other patterns which reduce in a way or an other the performance of the prediction. Youâ€™ve got to be careful, though, in general, dealing with categorical variables and random forests. Especially those with many levels. They not only increase the computational complexity to create the trees (and thus the forest), they can also take more importance than you would like. A quote from Wikipediaâ€™s article on random forests (in the variable selection section): â€œFor data including categorical variables with different number of levels, random forests are biased in favor of those attributes with more levels.â€
